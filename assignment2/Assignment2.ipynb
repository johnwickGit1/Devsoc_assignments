Continue from loading the preprocessed data. 


```python
import pickle

with open("processed_data.pkl", "rb") as f:
    X_encoded, y_encoded = pickle.load(f)

X_encoded.shape, y_encoded.shape
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_encoded,
    y_encoded,
    test_size=0.2,
    random_state=42,
    stratify=y_encoded
)

X_train.shape, X_test.shape

#now we make the test and train split before traning the classifier
```

```python
#training the first decision tree (this is before moving on to random forest.)
from sklearn.tree import DecisionTreeClassifier

# Create model
dt = DecisionTreeClassifier(random_state=42)

# Train model
dt.fit(X_train, y_train)
```

```python
#now we test the accuracy (real evaluation of the first tree), ie by making predictions
y_pred = dt.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report

print("Accuracy (Decision Tree):", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n")
print(classification_report(y_test, y_pred))
```

```python
#we move to random forest now:
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(
    n_estimators=100,
    random_state=42
)

rf.fit(X_train, y_train)

y_pred_rf = rf.predict(X_test)

from sklearn.metrics import accuracy_score, classification_report

print("Random Forest Accuracy:", accuracy_score(y_test, y_pred_rf))
print("\nClassification Report:\n")
print(classification_report(y_test, y_pred_rf))
```

Decision tree --> 85%
Random Forest --> 89% 

~5% jump.



```python
#now we move on to boosting (where a new model learns from the errors of the previous one)
from sklearn.ensemble import GradientBoostingClassifier

gb = GradientBoostingClassifier(
    n_estimators=100,
    learning_rate=0.1,
    random_state=42
)

gb.fit(X_train, y_train)

y_pred_gb = gb.predict(X_test)

from sklearn.metrics import accuracy_score, classification_report

print("Gradient Boosting Accuracy:", accuracy_score(y_test, y_pred_gb))
print("\nClassification Report:\n")
print(classification_report(y_test, y_pred_gb))
```

</h3> Now we start the voting implementation </h3>


```python
#1. Hard voting here we just see the class most models have predicted.
from sklearn.ensemble import VotingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.metrics import accuracy_score, classification_report

dt = DecisionTreeClassifier(random_state=42)
rf = RandomForestClassifier(n_estimators=100, random_state=42)
gb = GradientBoostingClassifier(random_state=42)

voting_hard = VotingClassifier(
    estimators=[("dt", dt), ("rf", rf), ("gb", gb)],
    voting="hard"
)

voting_hard.fit(X_train, y_train)
y_pred_hard = voting_hard.predict(X_test)

print("Hard Voting Accuracy:", accuracy_score(y_test, y_pred_hard))
print(classification_report(y_test, y_pred_hard))

#after running -> expected results.
```

```python
#soft voting -> average probablitites
voting_soft = VotingClassifier(
    estimators=[("dt", dt), ("rf", rf), ("gb", gb)],
    voting="soft"
)

voting_soft.fit(X_train, y_train)
y_pred_soft = voting_soft.predict(X_test)

print("Soft Voting Accuracy:", accuracy_score(y_test, y_pred_soft))
print(classification_report(y_test, y_pred_soft))
```

```python
#weighted soft voting -> gives influence to better models
voting_weighted = VotingClassifier(
    estimators=[("dt", dt), ("rf", rf), ("gb", gb)],
    voting="soft",
    weights=[1, 2, 3]   # dt, rf, gb
)

voting_weighted.fit(X_train, y_train)
y_pred_weighted = voting_weighted.predict(X_test)

print("Weighted Soft Voting Accuracy:", accuracy_score(y_test, y_pred_weighted))
print(classification_report(y_test, y_pred_weighted))
```

```python
# summary of voting:-
# dt : 0.845616570327553
# rf : 0.8978805394990366
# gb : 0.9063102119460501
# hard : 0.8995664739884393
# soft : 0.8798169556840078
# soft weighted : 0.9022157996146436

# expected, except for hard and soft which may sometimes differ.
```

```python
from sklearn.ensemble import StackingClassifier
from sklearn.linear_model import LogisticRegression

# Base models
dt = DecisionTreeClassifier(random_state=42)
rf = RandomForestClassifier(n_estimators=100, random_state=42)
gb = GradientBoostingClassifier(random_state=42)

stack = StackingClassifier(
    estimators=[
        ("dt", dt),
        ("rf", rf),
        ("gb", gb)
    ],
    final_estimator=LogisticRegression(max_iter=2000),
    cv=5
)

stack.fit(X_train, y_train)

y_pred_stack = stack.predict(X_test)

print("Stacking Accuracy:", accuracy_score(y_test, y_pred_stack))
print(classification_report(y_test, y_pred_stack))
```

```python
#lets implement blending now (does not have cv, like we had 5 fold cv in stacking)
#this uses a "test" data set from the already existing train data.

#step 1- creating the split
from sklearn.model_selection import train_test_split

X_train_blend, X_val_blend, y_train_blend, y_val_blend = train_test_split(
    X_train, y_train,
    test_size=0.2,
    random_state=42,
    stratify=y_train
)
```

```python
#step2 -> training the models on this created "test split"
dt.fit(X_train_blend, y_train_blend)
rf.fit(X_train_blend, y_train_blend)
gb.fit(X_train_blend, y_train_blend)
```

```python
#testing the model for the validation set(what the model created)
import numpy as np

val_preds = np.column_stack([
    dt.predict_proba(X_val_blend),
    rf.predict_proba(X_val_blend),
    gb.predict_proba(X_val_blend)
])

meta_model = LogisticRegression(max_iter=2000)
meta_model.fit(val_preds, y_val_blend)
```

```python
test_preds = np.column_stack([
    dt.predict_proba(X_test),
    rf.predict_proba(X_test),
    gb.predict_proba(X_test)
])

y_pred_blend = meta_model.predict(test_preds)

print("Blending Accuracy:", accuracy_score(y_test, y_pred_blend))
print(classification_report(y_test, y_pred_blend))
```

we see blending ~ stacking which is expected. we had so much data approx 20k rows so that stacking had a good enough validation dataset, to produce a good result.

```python
#before doing this expt -> rf uses no max depth by default (so i expect rf_shallow to not be high in accuracy)
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report

rf_shallow = RandomForestClassifier(
    n_estimators=200,
    max_depth=5,         
    random_state=42,
    n_jobs=-1
)

rf_deep = RandomForestClassifier(
    n_estimators=200,
    max_depth=None,       
    random_state=42,
    n_jobs=-1
)

# Train
rf_shallow.fit(X_train, y_train)
rf_deep.fit(X_train, y_train)

# Predict
y_pred_shallow = rf_shallow.predict(X_test)
y_pred_deep = rf_deep.predict(X_test)

# Evaluate
print("RF Shallow Accuracy:", accuracy_score(y_test, y_pred_shallow))
print(classification_report(y_test, y_pred_shallow))

print("\n")

print("RF Deep Accuracy:", accuracy_score(y_test, y_pred_deep))
print(classification_report(y_test, y_pred_deep))

#after expriment -> valid ~10% inc in accuracy.
```
